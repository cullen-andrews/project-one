```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_2018 = "Resources/clean2018.csv"
data_2018_df = pd.read_csv(data_2018)
data_2018_df.head()
```

```python
#UGDS_AIAN
csc_2018_df = data_2018_df.loc[:, ["STABBR","UGDS_WHITE", "UGDS_BLACK", "UGDS_ASIAN", "UGDS_HISP"]] 
csc_2018_df.head()
```

```python
average_race_2018 = csc_2018_df.mean(axis=0)
print(average_race_2018)
```

```python
data_2014 = "Resources/clean2014.csv"
data_2014_df = pd.read_csv(data_2014)
data_2014_df.head()
```

```python
#"UGDS_AIAN"
csc_2014_df = data_2014_df.loc[:, ["STABBR", "UGDS_WHITE", "UGDS_BLACK", "UGDS_ASIAN", "UGDS_HISP"]]
csc_2014_df.head()
```

```python
average_race_2014 = csc_2014_df.mean(axis=0)
print(average_race_2014)
```

```python
data_2011 = "Resources/clean2011.csv"
data_2011_df = pd.read_csv(data_2011)
data_2011_df.head()
```

```python
# "UGDS_AIAN"
csc_2011_df = data_2011_df.loc[:, ["STABBR", "UGDS_WHITE", "UGDS_BLACK", "UGDS_ASIAN" , "UGDS_HISP"]]
csc_2011_df.head()
```

```python
average_race_2011 = csc_2011_df.mean(axis=0)
print(average_race_2011)
```

```python
three_years_race_admits_df = pd.DataFrame({"2011":average_race_2011, "2014":average_race_2014,"2018":average_race_2018})

three_years_race_admits_df                 
```

```python
renamed_all_df = three_years_race_admits_df.rename(index={"UGDS_WHITE":"White",
                                          "UGDS_BLACK":"Black",
                                          "UGDS_ASIAN":"Asian", 
                                          "UGDS_HISP":"Hispanic or Latino"})
renamed_all_df
```

```python
df_style = renamed_all_df.style.format({"2011": "{:.2%}",
                                         "2014": "{:.2%}",
                                         "2018": "{:.2%}"
                                        })
    
display(df_style)
```

```python
renamed_all_df.plot.bar()

plt.title("Percent of Students, by Race, Admitted to Colleges in the U.S.")
plt.ylabel("Percent of Total Admitted Students")
plt.xlabel("Race")

plt.savefig ("race_of_admitted_students.png")

plt.show()
plt.tight_layout()
```
