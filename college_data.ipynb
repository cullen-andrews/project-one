```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_2018 = "Resources/clean2018.csv"
data_2018_df = pd.read_csv(data_2018)
data_2018_df.head()
```

```python
csc_2018_df = data_2018_df.loc[:, ["STABBR","UGDS_WHITE", "UGDS_BLACK", "UGDS_HISP", "UGDS_ASIAN", "UGDS_AIAN"]]
csc_2018_df.head()
```

```python
average_race_2018 = csc_2018_df.mean(axis=0)
print(average_race_2018)
```
