```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census_data = "cleaned_census_data2011-2014-2018.csv"
census_data_df = pd.read_csv(census_data)
# census_data_df.head()
census_data_df.columns
```

```python
census_data_income_df = census_data_df.loc[:, ["Name",
                                               "White Median Income_2011",
                                               "Black Median Income_2011", 
                                               "Asian Median Income_2011", 
                                               "Hispanic or Latino Median Income_2011", 
                                               "White Median Income_2014",
                                               "Black Median Income_2014", 
                                               "Asian Median Income_2014", 
                                               "Hispanic or Latino Median Income_2014", 
                                               "White Median Income",
                                               "Black Median Income", 
                                               "Asian Median Income", 
                                               "Hispanic or Latino Median Income"]]
census_data_income_df.head()
```

```python
income_2011_df = census_data_income_df.loc[:, ["Name",
                                               "White Median Income_2011",
                                               "Black Median Income_2011", 
                                               "Asian Median Income_2011", 
                                               "Hispanic or Latino Median Income_2011"]]

renamed_2011_df = income_2011_df.rename(columns={"White Median Income_2011":"White",
                                                "Black Median Income_2011":"Black",
                                                "Asian Median Income_2011":"Asian", 
                                               "Hispanic or Latino Median Income_2011":"Hispanic or Latino"})
renamed_2011_df.head()
```

```python
average_inc_race_2011 = renamed_2011_df.mean() #axis=0 needed?
print(average_inc_race_2011)
```

```python
income_2014_df = census_data_income_df.loc[:, ["Name",
                                               "White Median Income_2014",
                                               "Black Median Income_2014", 
                                               "Asian Median Income_2014", 
                                                "Hispanic or Latino Median Income_2014"]]

renamed_2014_df = income_2014_df.rename(columns={"White Median Income_2014":"White",
                                                "Black Median Income_2014":"Black",
                                                "Asian Median Income_2014":"Asian", 
                                               "Hispanic or Latino Median Income_2014":"Hispanic or Latino"})
renamed_2014_df.head()
```

```python
average_inc_race_2014 = renamed_2014_df.mean() #axis=0 needed?
print(average_inc_race_2014)
```

```python
income_2018_df = census_data_income_df.loc[:, ["Name",
                                               "White Median Income",
                                               "Black Median Income", 
                                               "Asian Median Income", 
                                               "Hispanic or Latino Median Income"]]

renamed_2018_df = income_2018_df.rename(columns={"White Median Income":"White",
                                                "Black Median Income":"Black",
                                                "Asian Median Income":"Asian", 
                                               "Hispanic or Latino Median Income":"Hispanic or Latino"})

renamed_2018_df.head()
```

```python
average_inc_race_2018 = renamed_2018_df.mean() #axis=0 needed?
print(average_inc_race_2018)
```

```python
three_years_race_income_df = pd.DataFrame({"2011":average_inc_race_2011, "2014":average_inc_race_2014,"2018":average_inc_race_2018})
three_years_race_income_df
```

```python
three_years_race_income_df.plot.bar()

plt.title("Average Annual Household Income, by Race in the U.S.")
plt.ylabel("Average Annual Household Income ($)")
plt.xlabel("Race")
plt.legend(bbox_to_anchor=(1.05, 1))

plt.savefig ("ave_household_inc_by_race.png")

plt.show()
plt.tight_layout()
```
